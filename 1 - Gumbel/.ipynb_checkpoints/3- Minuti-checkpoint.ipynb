# Metodo dei Momenti

```python
import os
import math
from pandas import *
import pandas as pd
from datetime import datetime
%matplotlib inline 
import matplotlib.pyplot as plt
from scipy.optimize import least_squares 
from scipy.stats import genextreme,gumbel_r
from numpy import linspace
from statsmodels.distributions.empirical_distribution import ECDF
import numpy as np
from pylab import plot,show,hist,figure,title
```

```python
data = pd.read_csv('PluviometriaMin.txt')
```

```python
map(datetime,data['anno'])
data=data.set_index('anno')
```

```python
data
```

```python
means=data.mean()
variances=data.var()
stds=data.std()
cv=data.std()/means
EulerGamma=0.577215664901532860606512090
```

```python
stds
```

```python
coeff1 = (math.sqrt(6)/math.pi)
coeff2= coeff1*EulerGamma
```

```python
def g(m,s):
    return pd.DataFrame([m-coeff2*s,coeff1*s],index=["a","b"])
```

```python
ab_m = g(means, stds)
ab_m
```

```python
ab_m.to_pickle("ab_mMin")
```

```python
def gumbel(x,abs_d,col):
    return np.exp(-np.exp(-(x-ab_m[col]["a"])/ab_m[col]["b"]))
```

```python
min= data.min()[0]
max = data.max()[2]
t_rain=np.linspace(min, max,100)
```

```python
gb=pd.DataFrame([gumbel(t_rain,ab_m,"15min"),
                 gumbel(t_rain,ab_m,"30min"),
                 gumbel(t_rain,ab_m,"45min")]).T
gb.index=t_rain
gb.columns=["15min","30min","45min"]
gb
```

```python
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10) 
plt.style.use('default')
ax=gb.plot(color=["red","blue","magenta","limegreen","orange"])
ecdf15 = ECDF(data["15min"])
ax.plot(data["15min"],ecdf15(data["15min"]),'+',c="red", label="15min",markersize=6) 
data30=data["30min"]
ecdf30 = ECDF(data30)
ax.plot(data30,ecdf30(data30),"+",c="blue", label="30min",markersize=6)
data45=data["45min"]
ecdf45 = ECDF(data45)
ax.plot(data45,ecdf45(data45),"+",c="magenta", label="45min",markersize=6)
ax.set_title('Curve di Gumbel')
ax.set_xlabel('Rainfall [mm]')
ax.set_ylabel('P[H < h]')
ax.set_aspect('auto')
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=0.5)
#minor grid lines
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=0.8)
plt.grid()
```

Per durate inferiori all'ora, le Curve di Gumbel sembrano rappresentare in modo più preciso i dati raccolti.
Si potrebbe spiegare questo fenomeno con il fatto che il valore delle deviazioni standard calcolate per durate di precipitazioni inferiori sono visibilmente minori rispetto a quelle calcolate per durate superiori all'ora.

# Metodo dei Minimi Quadrati

```python
def fun(x,t,y):
    return np.exp(-np.exp(-(t-x[0])/x[1]))-y
from scipy.optimize import least_squares
```

```python
ecdf15 = ECDF(data["15min"])
x0=[17.,4.9]
```

```python
t_train=sorted(data["15min"])
y_train=ecdf15(t_train)
```

```python
res_lsq_15=least_squares(fun,x0,args=(t_train,y_train))
res_lsq_15.x 
```

```python
ecdf30=ECDF(data['30min'])
t_rain=sorted(data['30min'].dropna())
y_train=ecdf30(t_train)
x0=[25,8.9]
res_lsq_30=least_squares(fun,x0,args=(t_train,y_train))
res_lsq_30.x
```

```python
ecdf45=ECDF(data['45min'])
t_rain=sorted(data['45min'].dropna())
y_train=ecdf45(t_train)
x0=[29.6,12.6]
res_lsq_45=least_squares(fun,x0,args=(t_train,y_train))
res_lsq_45.x
```

```python
ab_mqm=pd.DataFrame([res_lsq_15.x,res_lsq_30.x,res_lsq_45.x],
                    columns=['a','b'],
                    index=['15min','30min','45min']).T
ab_mqm
```

```python
ab_mqm.to_pickle("ab_mqMin")
```

```python
def gumbel(x,abs_mqm,col):
    return np.exp(-np.exp(-(x-ab_mqm[col]["a"])/ab_mqm[col]["b"]))

t_rain=np.linspace(data.min()[0],data.max()[2],100)
```

```python
gb=pd.DataFrame([gumbel(t_rain,ab_mqm,"15min"),
                 gumbel(t_rain,ab_mqm,"30min"),
                 gumbel(t_rain,ab_mqm,"45min")]).T
gb.index=t_rain
gb.columns=["15min","30min","45min"]
gb.head()
```

```python
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10) 
plt.style.use('default')
ax=gb.plot(color=["red","blue","magenta","limegreen","orange"])
ecdf15 = ECDF(data["15min"])
ax.plot(data["15min"],ecdf15(data["15min"]),'+',c="red", label="15min",markersize=6) 
data30=data["30min"]
ecdf30 = ECDF(data30)
ax.plot(data30,ecdf30(data30),"+",c="blue", label="30min",markersize=6)
data45=data["45min"]
ecdf45 = ECDF(data45)
ax.plot(data45,ecdf45(data45),"+",c="magenta", label="45min",markersize=6)
ax.set_title('Curve di Gumbel')
ax.set_xlabel('Rainfall [mm]')
ax.set_ylabel('P[H < h]')
ax.set_aspect('auto')
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=0.5)
#minor grid lines
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=0.8)
plt.grid()
```

# Test di Pearson

```python
ab_m=pd.read_pickle("ab_mMin")
ab_mq=pd.read_pickle("ab_mqMin")
ab_mle=pd.read_pickle("ab_mleMin")
```

```python
rv=gumbel_r(loc=ab_m.at["a","15min"],scale=ab_m.at["b","15min"])
```

```python
q=[0.2,0.4,0.6,0.8,1]
h1h=rv.ppf(q)
h1h
```

```python
def X2(data,abs_t,h,delta):    
    dt=data[h].dropna()
    lmin=len(dt)
    q=[delta*(i+1) for i in range(int(1/delta))]
    rv=gumbel_r(loc=abs_t.at["a",h],scale=abs_t.at["b",h])
    ecdf = ECDF(dt)
    r=ecdf(rv.ppf(q))
    o0=lmin*r
    o=o0-np.append([0],np.delete(o0,-1))
    e=[delta*lmin for i in range(len(r))]
    return ((o-e)**2/e).sum()
```

```python
a=[]
for  m in data.columns:
    a=a+[[X2(data,ab_m,m,0.2),X2(data,ab_mqm,m,0.2),X2(data,ab_mle,m,0.2)]]
   
Dati = pd.DataFrame(a,index=["15min","30min","45min"],columns=["M","Mq","MLE"])
Dati
```

```python
DatiT = Dati.T
DatiT.min()
```

```python
best=DataFrame([ab_mle["15min"],
                ab_mqm["30min"],
                ab_m["45min"]]).T
                  
best
```

## Linee di Possibilità Pluviometrica

```python
def qtls(df,qu):
    rv1h=gumbel_r(loc=df["a"],scale=df["b"])
    return rv1h.ppf(qu)
```

```python
def h_g(bs,col,tr):
    return bs[col]["a"]+bs[col]["b"]*(-math.log(-math.log(1-1/tr)))
```

```python
[h_g(best,"15min",10),h_g(best,"30min",10),h_g(best,"45min",10)]
```

```python
Tr=10
q10=1-1/Tr
q20=1-1/20
q100=1-1/100
```

```python
pts=DataFrame([[qtls(best["15min"],q10),
           qtls(best["30min"],q10),
           qtls(best["45min"],q10)],
          [qtls(best["15min"],q20),
           qtls(best["30min"],q20),
           qtls(best["45min"],q20)],
          [qtls(best["15min"],q100),
           qtls(best["30min"],q100),
           qtls(best["45min"],q100)]], index=[10,20,100],columns=[15,30,45]).T
pts
```

```python
ag=pts.plot(style="-o")
plt.xlim([0,90])
plt.ylim([25,90])
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=0.5)
#minor grid lines
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=0.8)
```

```python
def na(pt,tr):
    a=np.polyfit(np.log(pt.index),np.log(pt[tr]),1)
    return [a[0], np.exp(a[1])]
in10=na(pts,10)
in20=na(pts,20)
in100=na(pts,100)
```

```python
fnl=DataFrame([in10,in20,in100],index=[10,20,100],columns=["n","a"])
fnl
```

```python
def h(tp,a,n):
    return a*tp**n
```

```python
tp=np.linspace(10,70,100)
```

```python
h10=h(tp,fnl["a"][10],fnl["n"][10])
h20=h(tp,fnl["a"][20],fnl["n"][20])
h100=h(tp,fnl["a"][100],fnl["n"][100])
```

```python
inh=DataFrame([h10,h20,h100],index=["Tr = 10","Tr = 20","Tr = 100"],
              columns=tp).T
inh[-5:]
```

```python
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
plt.style.use('default')
ag=inh.plot(color= ["red","blue", "magenta" ])
ag.plot(pts[10],'+',c="red",label="10h", markersize=6)
ag.plot(pts[20],"+",c="blue",label="20h", markersize = 6)
ag.plot(pts[100],"+",c="magenta", label ="100h", markersize = 6)
#ag.set_yscale('log')
#ag.set_xscale('log')
plt.xlim([10,70])
plt.ylim([25,150])
ag.set_title('Curve di Possibilità Pluviometrica')
ag.set_xlabel('Durata [h]')
ag.set_ylabel('Precipitazinoe Cumulata [mm]')
ag.set_aspect('auto')
plt.legend(['10h', '20h', '100h'],  title = "Legenda", title_fontsize = 10)
#minor grid lines
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=0.8)
```
