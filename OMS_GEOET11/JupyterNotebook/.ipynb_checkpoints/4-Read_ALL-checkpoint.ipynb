```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
import os
```

```python
print(os.listdir('output/Evapotranspiration_PriestleyTaylor/'))
```

```python
dfElfao = pd.read_csv('output/Evapotranspiration_PriestleyTaylor/latentHeat_PT.csv',
                 skiprows=6, sep=',', parse_dates=[0], 
                 na_values=-9999,usecols=[1,2])
dfElfao.columns = ['Datetime','El pt']
dfElfao.index = dfElfao['Datetime']
dfElfao = dfElfao.iloc[:,1:]
dfElfao.head(3)
```

```python
dfElpt = pd.read_csv('output/Evapotranspiration_FAO/latentHeat_FAO.csv',
                 skiprows=6, sep=',', parse_dates=[0], 
                 na_values=-9999,usecols=[1,2])
dfElpt.columns = ['Datetime','El fao']
dfElpt.index = dfElpt['Datetime']
dfElpt = dfElpt.iloc[:,1:]
```

```python
dfElSun = pd.read_csv('output/TranspirationResistance/LatentHeat_SunStress.csv',
                 skiprows=6, sep=',', parse_dates=[0], 
                 na_values=-9999,usecols=[1,2])
dfElSun.columns = ['Datetime','El gf sun']
dfElSun.index = dfElSun['Datetime']
dfElSun = dfElSun.iloc[:,1:]
```

```python
dfElSh = pd.read_csv('output/TranspirationResistance/LatentHeat_ShadeStress.csv',
                 skiprows=6, sep=',', parse_dates=[0], 
                 na_values=-9999,usecols=[1,2])
dfElSh.columns = ['Datetime','El gf shade']
dfElSh.index = dfElSh['Datetime']
dfElSh = dfElSh.iloc[:,1:]
```

```python
dfElObs = pd.read_csv('data/El.csv',
                 skiprows=3, sep=',', parse_dates=[0], 
                 na_values=-9999,usecols=[1,2])
dfElObs.columns = ['Datetime','El obs']
dfElObs.index = dfElObs['Datetime']
dfElObs = dfElObs.iloc[:,1:]
```

```python
dfEl = pd.concat([dfElfao,dfElpt,dfElSun,dfElSh,dfElObs], sort=True, axis=1)
dfEl['El gf'] = dfEl['El gf sun'] + dfEl['El gf shade']
```

```python
dfEl.head(5)
```

```python
dfEl.plot(figsize=(50,5))
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(30,20))
ax1 = axes[0]
ax2 = axes[1] 
ax3 = axes[2] 

dfEl.plot(ax=ax1,y=['El fao','El obs'],kind='line', stacked=False,title='Penman Monteith Fao',
              label=['FAO','Obs'],linewidth=2)   

dfEl.plot(ax=ax2,y=['El pt','El obs'],kind='line', stacked=False,title='Priestley-Taylor',
              label=['PT','Obs'],linewidth=2) 
dfEl.plot(ax=ax3,y=['El gf','El obs'],kind='line', stacked=False,title='Geoframe',
              label=['GEOframe','Obs'],linewidth=2) 

for ax in [ax1, ax2, ax3]:
    ax.set_ylim((-100,750))
    ax.set_ylabel("Latent heat[W/m$^{2}$]",size = 20,color="k",alpha=1)
    ax.set_xlabel("Datetime",size = 20,color="k",alpha=1)   
    ax.tick_params(axis='both', which='major', labelsize=16)
    ax.tick_params(axis = 'both', which = 'minor', labelsize = 16)
    ax.legend(loc='upper left',fontsize=20)
    ax.title.set_size(30)
    
fig.add_subplot(111, frameon=False);
plt.grid(False)
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.tight_layout()    
```
