## Shortwave radiation

The equation representing shortwave radiation in a transparent atmosphere has the following expression:
\begin{equation}
S_{\downarrow} = C_{1}\cdot I_{sc} \cdot E_{0} \cdot cos(\theta_{s} ) \cdot (T_{s} + \beta_{s} ) \cdot \psi 
\end{equation}

 - $\theta_{s}$: angle between incident ray and surface;
 - $\Psi$: shading index.

```python
os.chdir(oms_project_path+'/data/Cavone/1')
nameSWFile='Cavone_ShortwaveDirect_1.csv'
```

```python
dfSw = pd.read_csv(nameSWFile,
                 skiprows=6, sep=',', parse_dates=[0], 
                 na_values=-9999,usecols=[1,2])
dfSw.columns = ['Datetime','SW']
dfSw.index = dfSw['Datetime']
dfSw = dfSw.iloc[:,1:]
dfSw.head(3)
```

```python
ax=dfSw.plot(color="crimson",figsize=(15,7))
ax.set_title('Shortwave Radiation',size=20)
ax.set_ylabel("SW [$W \cdot m^{−2}$]",size=15)
ax.set_xlabel("Datetime",size=15)
```

## Longwave radiation

Simplified models use the Stefan-Blotzmann relationship, considering both the atmosphere and the earth as two gray bodies:
\begin{equation}
L_{\downarrow} = \epsilon_{all-sky} · \sigma · T_{a}^{4} \\
L_{\uparrow}= \epsilon_{s} · \sigma · T_{s}^{4}
\end{equation}

\begin{equation}
\epsilon_{all−sky} = \epsilon_{clear} \cdot (1 + a \cdot c^{b})
\end{equation}

## Radiazione netta

Net radiation is a balance between incoming and outgoing radiation at a point on the Earth's surface; it is often divided into short wave and long wave.

$$ Rn=(1-\alpha)(Rs+d)+Rl_d-Rl_u $$

 - Rn: radiazione netta [$MJ \cdot m^{2} \cdot day^{−1}$]
 - $\alpha$: albedo
 - Rs: radiazione ad onda corta diretta [$MJ \cdot m^{2} \cdot day^{−1}$]
 - d: radiazione ad onda corta diffusa [$MJ \cdot m^{2} \cdot day^{−1}$]
 - Rl_d: radiazione ad onda lunga rilasciata dall'atmosfera [$MJ \cdot m^{2} \cdot day^{−1}$]
 - Rl_u: radiazione ad onda lunga rilascita dal suolo [$MJ \cdot m^{2} \cdot day^{−1}$]


```python
os.chdir(oms_project_path+'/data/Cavone/1')
nameFile='Cavone_Net_1.csv'
```

```python
ax=dfNet.plot(color="crimson",figsize=(15,7))
ax.set_title('Net Radiation',size=20)
ax.set_ylabel("NR [$W \cdot m^{−2}$]",size=15)
ax.set_xlabel("Datetime",size=15)
```
