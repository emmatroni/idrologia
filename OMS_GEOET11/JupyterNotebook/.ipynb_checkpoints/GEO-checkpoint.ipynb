La seguente modelizzazione rigurada soltanto la Traspirazione.
Gli input sono:
- Temperatura
- Radiazione ad onda corta
- Radiazione ad onda lunga
- Velocità del vento
- Umidità
- Contenuto d'acqua del terreno
- Pressione atmosferica
- Flusso di calore uscente dal suolo
- LAI

La grande differenza rispetto all'equazione di Penman-Monteith FAO è data dal fatto che in questo caso si tiene in considerazione la temperatura della foglia: essa cambia in funzione della radiazione incidente raggiungendo il valore di equililbrio termico.

Gli output sono:
- Flusso di calore lantente emesso sia dalle foglie di luce che quelle di ombra
- Flusso di calore sensibile emesso sia dalle foglie di luce che quelle di ombra
- Traspirazione totale
- Frazione di chioma al sole
- Radiazione di onda corta assorbita dalle foglie

Nella modellizzazione della traspirazione con Geoframe si considerano anche gli stress a cui una pianta può essere soggetta. Essi possono di tre tipi:

- **STRESS IDRICO**: minore è la disponibilità di acqua più l'organismo vegetale tende a trattenerla

```python
Image('Images/stress idrico.png')
```

- **STRESS legato alla RADIAZIONE**: la curva mostra un andamento non lineare tendente a 1

```python
     Image('Images/fotosintesi e apr.png')
```

- **STRESS legato alla TEMPERATURA**: in condizioni estreme di troppo caldo o troppo freddo gli stomi tendono a chiudersi.

Un ulteriore STRESS  da considerarsi potrebbe essere lagato all'umidità, tuttavia in questa esercitazione non se ne tiene conto.

Dal grafico si evidenzia che la traspirazione delle fragole risulta maggiore di quella dell'albero della gomma: ciò sembra in apparente contrasto con quanto trovato attraverso la modelizzazione sfruttata da FAO. In realtà bisogna considerare che i due modelli analizzano parametri di input diversi. La modelizzazione con GEOframe è più sofisticata in quanto si concentra sulla lamina fogliare, sull'optimum di temperatura e sulla tipologia di piantagione. L'albero della gomma, in quanto pianta arborea, presenta una chioma in paraziale ombra perciò la traspirazione delle foglie in questa zona sarà minima o nulla; la fragola invece, in quanto pianta erbacea, è completamente al sole.
Infine, il maggior apporto di traspirazione delle fragole si deve anche al range più ampio di temperatura nel quale riesce a mantenere le funzioni vitali.

## Temperatura

**confronto temperatura foglie al sole**

```python
tGfsun= pd.read_csv('output/TranspirationResistance/LeafTemperature_SunStressfragola.csv',
                 skiprows=6, sep=',', parse_dates=[0], 
                 na_values=-9999,usecols=[1,2])
tGfsun.columns = ['Datetime','Tsun fragola']
tGfsun.index = tGfsun['Datetime']
tGfsun= tGfsun.iloc[:,1:]
```

```python
tGgsun= pd.read_csv('output/TranspirationResistance/LeafTemperature_SunStressgomma.csv',
                 skiprows=6, sep=',', parse_dates=[0], 
                 na_values=-9999,usecols=[1,2])
tGgsun.columns = ['Datetime','Tsun gomma']
tGgsun.index = tGgsun['Datetime']
tGgsun= tGgsun.iloc[:,1:]
```

```python
dfDaysum = pd.concat([tGfsun,tGgsun], sort=True, axis=1)
ax=dfDaysum.plot(color=["navy","crimson"],figsize=(14,5),linewidth=0.7)
ax.tick_params(axis='both', which='major', labelsize=11)
ax.set_xlabel('Datetime', size=15)
ax.set_ylabel('Temperatura [K]', size=15)
ax.legend(fontsize=18, loc='upper right')
ax.set_title('Confronto Temperatura_foglie al sole', size=15)
```

Dal confronto della temperatura delle foglie al sole emerge che la tempertura delle foglie delle fragole risulta in genere più alta rispetto a quella delle foglie dell'albero di gomma.
Questo conferma l'analisi svolta precedentemente riguardo la traspirazione per i due diversi casi: maggiore è la temperatura fogliare, maggiore è la traspirazione.

L'analisi rigurdante la temperatura delle foglie all'ombra non viene eseguita in quanto mancherebbe uno dei termini di confronto poichè le piante di fragole non presentano parti ombreggiate.
