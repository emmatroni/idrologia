## **Penman-Monteith FAO**

La parametrizzazione di Penman-Monteith è impiegata dalla FAO per stimare l'evaporazione delle colture.
I parametri in input sono:
- Temperatura;
- Radiazione netta;
- Pressione atmosferica;
- Flusso di calore dal suolo;
- Umidità;
- Velocità del vento;
- Contenuto d'acqua;

In questa modellizazione dell'evapotraspirazione compaiono anche i valori di stress riferiti alla coltura. La simulazione si basa quindi sull'evaporazione reale che tiene in cosiderazione anche il contenuto d'acqua effettivo nel suolo, a differenza dell'evaporazione potenziale che non considera alcun limite sulla quantità d'acqua evapotraspirabile.


\begin{equation}
ET_{0}^{FAO} =  K_{s} \cdot K_{c}\cdot\dfrac{0.408\cdot \Delta\cdot (R_{n} - G)+\gamma\cdot u_{2}\cdot(e_{s}-e)\cdot\frac{C_{p}}{T+273}}{\Delta+\gamma\cdot(1+C_{d}\cdot u_{2})}
\end{equation}

- $ET_{0}$ or $ET$ evapotraspirazione [$mm \cdot day^{−1}$] or [$mm\cdot hour^{−1}$]; 
- $K_{s}$ coefficiente di stress idrico;
- $K_{c}$ coefficiente di coltura;
- $R_{n}$ radiazione netta [$MJ \cdot m^{2} \cdot day^{−1}$] or [$MJ \cdot m^{2}\cdot h^{−1}$]; 
- $G$ flusso di calore uscente dalla superficie [$MJ \cdot m^{2} \cdot day^{−1}$] or [$MJ \cdot m^{2}\cdot h^{−1}$]; 
- $T$ temperatura media giornaliera (°C) ; 
- $u_{2}$ velocità del vento ($m\cdot s^{−1}$); 
- $e_{s}$ media della pressione di vapore a saturazione($kP_{a}$); 
- $e$ valore attuale della pressione di vapore; 
- $\Delta$ variazione della pressione di vapore alla saturazione rispetto la temperatura  ($kP_{a}\dot °C^{−1}$); 
- $\gamma$ costante psicrometrica ($kP_{a} \cdot$ °$C^{−1}$); 
- $C_{d}$ = 0.34 and $C_{p}$  =100 in caso di serie temporale giornaliera.

Gli output della simuazione sono due serie temporali: 
 -  Calore latente [ (W/m$^{2}$)];
 -  Evapotraspirazione (calore latente convertito in mm per ora o per giorno).

## Confronto coltura di fragole e coltura di albero della gomma

```python
dfsum = pd.concat([dff,dfg], sort=True, axis=1)
ag=dfsum.plot(color=["navy","crimson"],figsize=(15,5), linewidth=1)
ag.set_title('Et PM_GEO',size=15)
ag.set_ylabel("ET [$mm\cdot hour^{−1}$]",size=15)
ag.set_xlabel("Daytime",size=15)
```

Dal grafico si nota che l'evapotraspirazione di un campo di fragole risulta inferiore a quello di un campo di alberi della gomma. Ciò trova spiegazione in funzione dei parametri dati in input: tengono conto della profondità delle radici e dell crop coefficient Kc, diversi per le rispettive colture.
L'albero della gomma essendo una pianta arborea è in grado di prelevare una quantità d'acqua maggiore rispetto a quella delle fragole, pianta erbacea (radici meno profonde).

```python
dfDaysum = pd.concat([dffDay,dfgDay], sort=True, axis=1)
ax=dfDaysum.plot(color=["navy","crimson"],figsize=(14,5.5))
ax.legend(fontsize=12, loc='upper right')
ax.set_title('Confronto  Et 11 luglio 2006',size=15)
ax.set_ylabel("ET [$mm\cdot hour^{−1}$]",size=15)
ax.set_xlabel("Hours",size=15)
```
