## Radiazione

L'evapotraspirazione non potrebbe avvenire senza un'apporto di radiazione alla superficie evapotraspirante: l'acqua per passare dallo stato liquido a quello gassoso, necessita di energia.
Per un corretto studio della radiazione incidente su una superficie serve un'analisi rigurdante:
- Inquadramento geomorfologico dell'area di studio e sky view factor
- Radiazione ad onda corta
- Radiazione ad onda lunga
- Radiazione netta

```python
raster_path = 'output/Radiation/skyviewRaster.asc'
with open(raster_path, 'r') as raster:
    raster_header = raster.readlines()[:6]
raster_header = [item.strip().split()[-1] for item in raster_header]
raster_cols = int(raster_header[0])
raster_rows = int(raster_header[1])
raster_xll = float(raster_header[2])
raster_yll = float(raster_header[3])
raster_cs = float(raster_header[4])
raster_nodata = float(raster_header[5])
raster_array = np.loadtxt(raster_path, dtype=np.float, skiprows=6)
raster_array[raster_array == raster_nodata] = np.nan
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.set_title('Skyview factor')
raster_extent = [
    raster_xll, raster_xll + raster_cols * raster_cs,
    raster_yll, raster_yll + raster_rows * raster_cs]
img_plot = ax.imshow(raster_array, extent=raster_extent, cmap='jet')
cbar = plt.colorbar(img_plot, orientation='vertical', shrink=0.5, aspect=14)
plt.style.use('classic')
cbar.set_label('')
plt.show()
```

## Shortwave radiation

L'equazione rappresentante la radiazione ad onda corta, in un'atmosfera trasparente, ha la seguente espressione:
\begin{equation}
S_{\downarrow} = C_{1}\cdot I_{sc} \cdot E_{0} \cdot cos(\theta_{s} ) \cdot (T_{s} + \beta_{s} ) \cdot \psi 
\end{equation}

dove:
 - $\theta_{s}$: angolo tra raggio incidente e superficie;
 - $\Psi$: indice di ombreggiamento.

Per la simulazione è necessario specificare le coordinte della stazione e la data del rilevamento.

## Longwave radiation

I modelli semplificati sfruttano la relazione di Stefan-Blotzmann, considerando sia l'atmosfera che la terra due corpi grigi:
\begin{equation}
L_{\downarrow} = \epsilon_{all-sky} · \sigma · T_{a}^{4} \\
L_{\uparrow}= \epsilon_{s} · \sigma · T_{s}^{4}
\end{equation}

\begin{equation}
\epsilon_{all−sky} = \epsilon_{clear} \cdot (1 + a \cdot c^{b})
\end{equation}

## Radiazione netta

La radiazione netta è un bilancio tra la radiazione entrante e quella uscente in un punto della superficie terrestre; spesso è suddivisa in onda corta e onda lunga.

$$ Rn=(1-\alpha)(Rs+d)+Rl_d-Rl_u $$

 - Rn: radiazione netta [$MJ \cdot m^{2} \cdot day^{−1}$]
 - $\alpha$: albedo
 - Rs: radiazione ad onda corta diretta [$MJ \cdot m^{2} \cdot day^{−1}$]
 - d: radiazione ad onda corta diffusa [$MJ \cdot m^{2} \cdot day^{−1}$]
 - Rl_d: radiazione ad onda lunga rilasciata dall'atmosfera [$MJ \cdot m^{2} \cdot day^{−1}$]
 - Rl_u: radiazione ad onda lunga rilascita dal suolo [$MJ \cdot m^{2} \cdot day^{−1}$]
