This notebook is meant to analyse the output of the calibration.


```python
import os
import pandas as pd
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.style as style 
import matplotlib.colors as cl
import matplotlib.dates as mdates
import matplotlib
%matplotlib inline

from hydroeval import *

from geoframepy.timeseries import io_csv

import warnings
warnings.filterwarnings('ignore')

style.available
style.use('seaborn-whitegrid')

nice_fonts = {
        "legend.frameon": True,
        "legend.fancybox": True,
        "legend.facecolor": "white",
        "axes.edgecolor": "0.8",
        "axes.linewidth": 0.6,
        "grid.linewidth":0.4,
        "text.usetex": True,
        "font.family": "serif",
        "axes.labelsize": 16,
        "font.size": 16,
        "legend.fontsize": 14,
        "xtick.labelsize": 14,
        "ytick.labelsize": 14,
}

my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ["#ece7f2","#d0d1e6","#a6bddb","#74a9cf","#3690c0","#0570b0","#045a8d","#023858"])
colors = ['#a6cee3', '#1f78b4', '#b2df8a', '#33a02c', '#fb9a99', '#e31a1c', '#fdbf6f', '#ff7f00']

plt.rcParams.update(nice_fonts)

project_path = os.path.dirname(os.getcwd())
```

## Read observed data

```python
obs = io_csv.pandas_read_OMS_timeseries(project_path+"/data/Timeseries/ex06_calibration_observed_theta.csv")
obs.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10/1.618)) 

ax.plot(obs['0'], label='-0.25 [m]', color=colors[0])
ax.plot(obs['1'], label='-0.5 [m]', color=colors[1])
ax.plot(obs['2'], label='-1.0 [m]', color=colors[2])
ax.plot(obs['3'], label='-1.5 [m]', color=colors[3])


ax.set_ylabel('$\\theta$ [-]')
ax.set_xlabel('Time [mm-dd hh:mm]')
ax.set_title('Observed water content')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
ax.xaxis.set_tick_params(rotation=45)

ax.legend()
plt.show()
```

## Read simulated data

```python
sim = xr.open_dataset(project_path+'/output/ex06_calibrated_0000.nc')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10/1.618)) 

ax.plot(obs['0'], label='-0.25 [m]', color=colors[0])
sim.theta.sel(depth=-0.25, method='nearest').plot(ax=ax, color=colors[1])


ax.set_ylabel('$\\theta$ [-]')
ax.set_xlabel('Time [mm-dd hh:mm]')
ax.set_title('Obs vs sim water content at -0.25 [m]')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
ax.xaxis.set_tick_params(rotation=45)

ax.legend()
plt.show()
```

## Index of goodness

### RMSE

```python
rmse(obs['0'].values, sim.theta.sel(depth=-0.25, method='nearest').values)
```

### KGE

```python
kge(obs['0'].values, sim.theta.sel(depth=-0.25, method='nearest').values)[0]
```

### NSE

```python
nse(obs['0'].values, sim.theta.sel(depth=-0.25, method='nearest').values)
```
