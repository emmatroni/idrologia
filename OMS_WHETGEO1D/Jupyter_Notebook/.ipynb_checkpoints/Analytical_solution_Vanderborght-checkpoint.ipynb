### Analytical solution

```python
# Analytical solution
# Credit:
# https://github.com/CH-Earth/laughTests
#
#
from scipy.optimize import newton

def k_psi(psi, ksat, alpha, n, m):
    if psi < 0:
        return ksat * ( 
            ( ( (1 - (psi*alpha)**(n-1) * (1 + (psi*alpha)**n)**(-m))**2) /  
              ( (1 + (psi*alpha)**n)**(m/2))))
    else:
        return ksat

    
def solve_psi(psi, q, ksat, alpha, n, m):
    return q - k_psi(psi, ksat, alpha, n, m)


def integrate_psi(psi0, psi1, qIn, ksat, alpha, n, m):
    psi = np.linspace(psi0, psi1, num=100)
    dpsi = psi[1] - psi[0]
    return -np.sum([dpsi / ((qIn / k_psi(p, ksat, alpha, n, m)) - 1) for p in psi])


def vanderborght_analytic(top_soil, bottom_soil, q=0.5, 
                          max_iters=100, delx_max=2., tol=1e-12, delpsi_scale=0.3):
    psi_init = -25.
    ksat, alpha, n, m = param_dict[bottom_soil]
    psi_bot = newton(solve_psi, psi_init, args=(q, ksat, alpha, n, m), maxiter=500)
    
    ksat, alpha, n, m= param_dict[top_soil]
    psi_top = newton(solve_psi, psi_init, args=(q, ksat, alpha, n, m), maxiter=500)
    
    psi = np.linspace(psi_bot, psi_top, num=500)
    depths = np.zeros_like(psi)
    depths[0] = 50.
    
    for i, (psi0, psi1) in enumerate(zip(psi[:-1], psi[1:])):
        psi0_base = psi0
        psi1_base = psi1
        dpsi = psi1_base - psi0
        depths[i + 1] = depths[i]
        
        for trial in range(max_iters):
            psi1 = psi0 + dpsi
            if dpsi > 0:
                psi1 = max(psi1, psi1_base)
            else:
                psi1 = min(psi1, psi1_base)
            dpsi = psi1 - psi0
            
            delx = integrate_psi(psi0, psi1, q, ksat, alpha, n, m)
            if np.abs(delx) > delx_max:
                dpsi *= delpsi_scale
            else:
                psi0 = psi1
                depths[i+1] += delx
            if abs(psi0 - psi1_base) < tol:
                break
    return psi, depths
```

```python
param_dict = {
    # Soil - ksat[cm d-1] , alpha [cm-1],   n,  m
    'sand': (1002.23, -0.15, 3.0, 1-1/3.0),
    'loam': (  50., -0.04, 1.6, 1-1/1.6),
    'clay': (  10.0224, -0.01, 1.1, 1-1/1.1)
}
```

```python
line_w = 0.7
marker_s = 1.3


plot_width = 6


fig, ax = plt.subplots(1, 3, sharey='row', figsize=(plot_width,plot_width/1.618))
plt.plot(np.nan,np.nan, linewidth=linewidth_n, color='red', label='Numerical sol.')
plt.plot(np.nan,np.nan, linewidth=linewidth_n, linestyle='--', color='0.2', label='Analytical sol.')

# clay sand
ax[0].plot(ds1.psi.sel(time='2022-01-01 00:00').values, ds1.depth, color='red', linewidth=linewidth_n)

psi, depth = vanderborght_analytic('clay', 'sand')
ax[0].plot(psi[-depth<0]/100, -depth[-depth<0]/100, color='0.2', linestyle='--', linewidth=linewidth_a)
psi_bot = psi[np.argmax(depth)]
ax[0].plot([psi_bot/100, psi_bot/100], [-50/100, -200/100],  color='0.2', linestyle='--', linewidth=linewidth_a)
if min(depth) > 0:
    psi_surf = psi[np.argmin(depth)]
    ax[0].plot([psi_surf/100, psi_surf/100], [0, min(depth)], color='0.2', linestyle='--', linewidth=linewidth_a)


ax[0].set_ylabel('Depth [m]')
ax[0].set_xlabel('$\psi$ [m]')

#loam sand
ax[1].plot(ds2.psi.sel(time='2022-01-01 00:00').values, ds2.depth, color='red', linewidth=linewidth_n)

psi, depth = vanderborght_analytic('loam', 'sand')
ax[1].plot(psi[-depth<0]/100, -depth[-depth<0]/100, color='0.2', linestyle='--', linewidth=linewidth_a)
psi_bot = psi[np.argmax(depth)]
ax[1].plot([psi_bot/100, psi_bot/100], [-50/100, -200/100],  color='0.2', linestyle='--', linewidth=linewidth_a)
# psi_surf = psi[np.argmin(depth)]
# ax[1].plot([psi_surf/100, psi_surf/100], [0, max(-depth/100)], color='0.2', linestyle='--', linewidth=linewidth_a)


ax[1].set_xlabel('$\psi$ [m]')


# sand loam
ax[2].plot(ds3.psi.sel(time='2022-01-01 00:00').values, ds3.depth, color='red', linewidth=linewidth_n)

psi, depth = vanderborght_analytic('sand', 'loam')
ax[2].plot(psi/100, -depth/100, color='0.2', linestyle='--', linewidth=linewidth_a)
psi_bot = psi[np.argmax(depth)]
ax[2].plot([psi_bot/100, psi_bot/100], [-50/100, -200/100],  color='0.2', linestyle='--', linewidth=linewidth_a)
psi_surf = psi[np.argmin(depth)]
ax[2].plot([psi_surf/100, psi_surf/100], [0, max(-depth/100)], color='0.2', linestyle='--', linewidth=linewidth_a)

    
ax[2].set_xlabel('$\psi$ [m]')


ax[0].annotate('(a)', (0.1,0.9), xycoords='axes fraction', fontsize=10)
ax[1].annotate('(b)', (0.8,0.9), xycoords='axes fraction', fontsize=10)
ax[2].annotate('(c)', (0.1,0.9), xycoords='axes fraction', fontsize=10)

plt.legend(bbox_to_anchor=(0.5,1.2), ncol=2)
plt.show()
```
