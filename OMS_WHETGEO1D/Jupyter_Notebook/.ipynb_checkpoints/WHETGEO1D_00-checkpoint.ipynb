```python
ds_c = xr.open_dataset(r'C:\Users\Niccolo\OMS\OMS_Project_WHETGEO1D_calibration\output\Sim_a_1_layer_calibrated_thetas_alpha_ks_long_0000.nc')
```

```python
ds.theta.sel(time=slice('2003-01-01 00:00', '2003-03-01 00:00')).sel(depth=-0.25,method='nearest').plot(label='obs')
ds_c.theta.sel(time=slice('2003-01-01 00:00', '2003-03-01 00:00')).sel(depth=-0.25,method='nearest').plot(label='sim')
# plt.hlines(0.41, xmin=ds.time.values[0],xmax='2003-03-01 00:00')
plt.ylabel('$\\theta$ [-]')
plt.ylim([0.095,0.41])
plt.legend()
```
